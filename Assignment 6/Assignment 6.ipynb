(a) Compute function for each individual (map)

<font size = "3">

- Import the datset "data/investment.csv"
- Write a lambda function for $V = P/(1+r)$.
- Use ```list(map(...))``` to calculate $V$ for each row of the dataset


````python
# Write your own code here
investments = pd.read_csv("data/investment.csv")

compute_lambda = lambda P, r: P / (1 + r)
investments["V"] = list(map(compute_lambda,
                            investments["P"],
                            investments["r"]))

display(investments)

````

(c) Extract sorted elements from a dataset

<font size = "3">

- Sort by "prop_urbanpopulation" in descending order
- Subset the 5 countries with the highest urban population

````python
# Write your code here
wdi_2020 = pd.read_csv("data/wdi_2020.csv")
wdi_2020_sorted = wdi_2020.sort_values(by = "prop_urbanpopulation", ascending = False)
wdi_2020_sorted_subset = wdi_2020_sorted.iloc[0:5]
display(wdi_2020_sorted_subset)
````

````python
# Write your own code here
df = wdi_2020["prop_urbanpopulation"]
median = df.median(skipna = True)

above_median = wdi_2020.query("prop_urbanpopulation > @median")
display(above_median)
````

(e) Plot by category

<font size = "3">

- Read "data/wdi_urban.csv"<br>
This contains data from 1980 and 2020

- Create a list with the unique values of "year"
- Create a plot with a ...
    - Histogram of "prop_urbanpopulation" <br>
    for each year
    - Set the histogram option ``` alpha = 0.5 ``` for transparency
    - All the histograms should be overlapping
    - Create a legend with the list of years
    - Label the axes and the title

What do we learn from the graph?

HINT: See the example at the end of Lecture 10


````python
# Write your own code here
wdi_urban = pd.read_csv("data/wdi_urban.csv")
years_unique = pd.unique(wdi_urban["year"])
# print(years_unique)

for year in years_unique:
    temp_df = wdi_urban.query("year == @year")
    plt.hist(temp_df["prop_urbanpopulation"], alpha = 0.5)

plt.title("Prop Urban Population by Year")
plt.xlabel("Prop Urban Population")
plt.ylabel("Frequency")
plt.legend(labels = years_unique)
plt.show()
````
